```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


df = pd.read_csv('../Data/titanic.csv')
df
```

```python
df.corr()
```

```python
#Bar chart with the amount of survivors based on class.
#Easy to see that the first class suvived more, compared to third class.
sns.barplot(x = 'Pclass', y = 'Survived', data = df)
plt.savefig('../Results/survivors based on class.png')
```

```python
#Bar chart for Survived and gender.
sns.barplot(x = 'Sex', y = 'Survived', data = df)
plt.savefig('../Results/Survived and gender.png')
```

```python
#Bar chart for age and survivers.
#Age was not a really big deal while saving the passenger.
sns.barplot(x = 'Survived', y = 'Age', data = df)
plt.savefig('../Results/Age and survivers.png')
```
